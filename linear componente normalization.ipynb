```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import numpy as np
import matplotlib.pyplot as plt
```

```python
data_path = '../train_1.csv/train_1.csv'
import csv


def data_rows(filepath):
    with open(data_path) as csvfile:
        spamreader = csv.reader(csvfile)
        
        #next(spamreader) #skip first row        
        
        while(True):
            try:
               next_value = next(spamreader)
            except:
                continue

            if next_value == None:
                break
            
            yield next_value
```

```python
def get_page_info(page):
    underscore_split = page.split("_")
    
    agent = underscore_split[-1]
    access_type = underscore_split[-2]
    pagelang = underscore_split[-3].split(".")[0]
    pagename = "_".join(underscore_split[:-3])
    
    return pagename, pagelang, access_type, agent
```

```python
def get_line_func(y_data):
    x_data = range(len(y_data))
    poly = np.polyfit(x_data, y_data, 1)
    line_func = np.poly1d(poly)
    return line_func
```

```python
d_rows = data_rows(data_path)
first_row = next(d_rows) #skip first line  

#print(first_row[491])
    
start_index = 0
stop_index = 20
for i, r in enumerate(d_rows):
    if i < start_index:
        continue
    try:  
        %time project_data(r)
        print()
    
    except Exception as e:
        print(e)
        
    if i >= stop_index: 
        break
```
